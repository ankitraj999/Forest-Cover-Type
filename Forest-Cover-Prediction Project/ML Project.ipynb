```python
col_names = []
for i in xrange(0, len(df.columns)-1):
    col_names.append('f'+str(i))
col_names.append('label')
```

```python
#converting panda dataframe into h2o dataframe
```

```python
h2o_df = h2o.H2OFrame(df)
```

```python
h2o_df.shape
```

```python
#load the data and prepare for the modeling
```

```python
#generate random number and create training and validation
```

```python
train , valid = h2o_df.split_frame()
```

```python
train.shape
```

```python
valid.shape
```

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
x = train.names[:-1]#prediction columns
y = train.names[-1]#response columns
```

```python
# Encode response columns as categorical for multinomial classification
train[y] = train[y].asfactor()
valid[y] = valid[y].asfactor()
```

```python
#Train deep learning model
```

```python
model = H2ODeepLearningEstimator(
distribution="multinomial",
activation="RectifierWithDropout",
hidden=[32,32,32],
input_dropout_ratio=0.3,
sparse=True,
l1=1e-5,
epochs=100)
```

```python
model.train(
x=x,
y=y,
training_frame=train,
validation_frame=valid)
```

```python
model
```

```python
#now mse is 0.37 ,tweak the models
```

```python
hidden_opt = [[32,32],[32,16,8],[100]]
l1_opt = [1e-4,1e-3]
hyper_parameters = {"hidden":hidden_opt, "l1":l1_opt}
```

```python
#import h20 grid search
```

```python
from h2o.grid.grid_search import H2OGridSearch
model_grid = H2OGridSearch(H2ODeepLearningEstimator,
hyper_params=hyper_parameters)
model_grid.train(x=x, y=y,
distribution="multinomial", epochs=1000,training_frame=train, validation_frame=valid,
score_interval=2, stopping_rounds=3,
stopping_tolerance=0.05,
stopping_metric="misclassification")
```

```python
#print model grid search result
```

```python

for model in model_grid:
    print model.model_id + " mse: " + str(model.mse(valid=True))
    print model
```

```python
model_grid
```

```python
#forumulate model_ids and corresponding model mse in dictionary form
```

```python
model_dic = dict((str(model.model_id), model.mse(valid=True)) for model in model_grid)
```

```python
#finding model with min. mse value
```

```python
model_dic
```

```python
min_mse = min(model_dic.values())
```

```python
min_mse
```

```python
#finding best model
```

```python
for model in model_grid:
    if model.mse(valid=True)== min_mse:
        model_path = h2o.save_model(model, force=True)
        best_model = model
```
